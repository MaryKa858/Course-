**Задание 1:**

Используя класс `scipy.stats.poisson` самостоятельно проанализируйте **распределение Пуассона, имеющего λ = 14.**

--Рассчитайте параметры, описывающие распределение.

--Постройте график с гистограммой относительных частот.

--Постройте график с кумулятивной функцией распределения.

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import poisson
from scipy.stats import mstats
```

```python
#создание датасета

mean = 14 #λ = 14
size = 30
data = poisson.rvs(mu = mean, size = size)
print(data)
```

```python
# расчет статистических параметров распределения

res = mstats.describe(data)

print(f"Количество наблюдений: {res.nobs}\n"
      f"Мин: {res.minmax[0]:.2f}\n"
      f"Mакс: {res.minmax[1]:.2f}\n"
      f"Выборочное среднее: {res.mean:.2f}\n"
      f"Дисперсия: {res.variance:.2f}\n"
      f"Стандартное отклонение: {np.sqrt(res.variance):.2f}\n"
      f"Коэф. эксцесса: {res.kurtosis:.2f}\n"
      f"Асимметрия: {res.skewness:.2f}")
```

```python
from statistics import mode  
result = mode(data)  
print("Мода списка: ", result)
```

```python
# график гистрограммы относительных частот 

plt.hist(data, bins = 10, color = 'skyblue', edgecolor = 'black')

plt.title('Гистограмма относительных частот', fontsize=14, fontweight='bold')
plt.xlabel('Значения', fontsize=12)
plt.ylabel('Относительная частота', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
#график с кумулятивной функцией распределения

sort_data = np.sort(data)
y = np.linspace(0,1, len(data), endpoint = False)
plt.plot(sort_data, y)
plt.title('CDF')
plt.xlabel('Значения')
plt.ylabel('CDF')
plt.grid(True)
plt.show()
```

```python
from scipy import stats

# Сортировка данных для построения CDF
sorted_data = np.sort(data)

# Вычисление кумулятивных вероятностей
n = len(sorted_data)
cumulative_prob = np.arange(1, n + 1) / n

# Построение графика
plt.figure(figsize=(12, 6))

# Ступенчатый график CDF
plt.step(sorted_data, cumulative_prob, where='post', 
         color='red', linewidth=2.5, label='Эмпирическая CDF')

# Дополнительные точки для лучшей визуализации
plt.plot(sorted_data, cumulative_prob, 'o', markersize=4, 
         color='darkred', alpha=0.7)

# Настройки графика
plt.title('Кумулятивная функция распределения (CDF)', 
          fontsize=14, fontweight='bold')
plt.xlabel('Значения данных', fontsize=12)
plt.ylabel('Вероятность', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(0, 1.05)
plt.xlim(min(data) - 1, max(data) + 1)

# Добавление информации
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()
```

**1. Автобусы идут с интервалом 5 минут. Полагая, что случайная величина Х - время ожидания автобуса на остановке – распределена равномерно на указанном интервале, найдите:**

среднее время ожидания;

среднеквадратическое отклонение времени ожидания.

```python
from scipy.stats import uniform
import numpy as np

a,b = 0,5

# Создание равномерного распределения
uniform_dist = uniform(loc=a, scale=b-a)

# Среднее время ожидания
mean_waiting = uniform_dist.mean()
print(f"Среднее время ожидания: {mean_waiting:.2f} минуты")

# Среднеквадратическое отклонение
std_waiting = uniform_dist.std()
print(f"Среднеквадратическое отклонение: {std_waiting:.4f} минуты")

# Дисперсия
variance_waiting = uniform_dist.var()
print(f"Дисперсия: {variance_waiting:.4f}")
```

Ответ: а) 2.50, б) 1.4434

**2. Регистратор МФЦ (многофункциональный центр) обслуживает в среднем 3 заявителя в час. Найдите:**

закон распределения случайной величины Х – количество обслуженных заявителей в час;

вероятность того, что за 2 часа будет обслужено 8 заявителей;

вероятность того, что за 2 часа будет обслужено более 10 заявителей.

```python
mu_hour = 3 # 3 заявителя в час

# 1. Закон распределения случайной величины X (количество заявителей в час)
print("1. ЗАКОН РАСПРЕДЕЛЕНИЯ (за 1 час):")
print("k\tP(X = k)")

k_values = np.arange(0, 11)  # от 0 до 10 заявителей
for k in k_values:
    probability = poisson.pmf(k, mu_hour)
    print(f"{k}\t{probability:.4f}")

# 2. Вероятность того, что за 2 часа будет обслужено 8 заявителей
mu_2hours = mu_hour * 2  # 6 заявителей за 2 часа
prob_8_in_2h = poisson.pmf(8, mu_2hours)

print(f"\n2. Вероятность обслужить 8 заявителей за 2 часа:")
print(f"Параметр за 2 часа: μ = {mu_hour} × 2 = {mu_2hours}")
print(f"P(X = 8) = {prob_8_in_2h:.4f}")

# 3. Вероятность того, что за 2 часа будет обслужено более 10 заявителей
prob_gt_10_in_2h = 1 - poisson.cdf(10, mu_2hours)

print(f"\n3. Вероятность обслужить более 10 заявителей за 2 часа:")
print(f"P(X > 10) = 1 - P(X ≤ 10) = 1 - {poisson.cdf(10, mu_2hours):.4f} = {prob_gt_10_in_2h:.4f}")
```

Ответ: 1) Пуассон 2) 0.1033 3) 0.0426

**3. Вероятность опоздания поезда на станцию более, чем на 5 минут составляет 5%. В ближайшее время на станцию должно приехать 5 поездов.** Определите:

закон распределения случайной величины Х – число опозданий поезда более, чем на 5 минут;

вероятность того, что ни один поезд не опоздает;

вероятность того, что опоздают ровно 2 поезда;

вероятность того, что опоздает хотя бы два поезда.

```python
n,p = 5, 0.05
mu = n * p  # λ = n × p = 5 × 0.05 = 0.25

print("\n" + "="*50)
print("=== РАСПРЕДЕЛЕНИЕ ПУАССОНА (приближенное решение) ===")
print(f"X ~ Pois(λ={mu})")
print("Примечание: Пуассон приближает биномиальное при малых p")

# Закон распределения
print("\n1. ЗАКОН РАСПРЕДЕЛЕНИЯ:")
print("k\tP(X = k)")

for k in range(0, 6):  # от 0 до 5
    probability = poisson.pmf(k, mu)
    print(f"{k}\t{probability:.6f}")

# 2. Вероятность того, что ни один поезд не опоздает
prob_0_pois = poisson.pmf(0, mu)
print(f"\n2. Вероятность, что ни один поезд не опоздает:")
print(f"P(X = 0) = {prob_0_pois:.6f}")

# 3. Вероятность того, что опоздают ровно 2 поезда
prob_2_pois = poisson.pmf(2, mu)
print(f"\n3. Вероятность, что опоздают ровно 2 поезда:")
print(f"P(X = 2) = {prob_2_pois:.6f}")

# 4. Вероятность того, что опоздает хотя бы два поезда
prob_ge_2_pois = 1 - poisson.cdf(1, mu)
print(f"\n4. Вероятность, что опоздает хотя бы два поезда:")
print(f"P(X ≥ 2) = 1 - P(X ≤ 1) = 1 - {poisson.cdf(1, mu):.6f} = {prob_ge_2_pois:.6f}")
```

Ответ: 1)Пуассона 2)0.778801 3)0.024338 4)0.026499
